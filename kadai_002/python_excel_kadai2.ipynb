```python
def performance(sales, average_sales):
    """
    Calculate the performance rank based on sales and average sales.

    Args:
        sales (float): The individual sales figure.
        average_sales (float): The overall average sales figure.

    Returns:
        str: The performance rank ('A', 'B', or 'C').
    """
    if sales >= average_sales + 50:
        return 'A'
    elif sales >= average_sales:
        return 'B'
    else:
        return 'C'

# Test the function with some examples
print(f"Sales 220, Average 200: {performance(220, 200)}") # Expected A (220 >= 200 + 50 = 216.67)
print(f"Sales 180, Average 200: {performance(180, 200)}") # Expected B (180 >= 200)
print(f"Sales 100, Average 200: {performance(100, 200)}") # Expected C (100 < 200)
print(f"Sales 216.67, Average 200: {performance(216.67, 166.67)}") # Expected A (Boundary case for A)
print(f"Sales 200, Average 200: {performance(200, 200)}") # Expected B (Boundary case for B)
```
