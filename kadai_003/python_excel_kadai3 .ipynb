```python
import pandas as pd
import openpyxl
from openpyxl.styles import PatternFill

output_file = '売上集計表.xlsx'

# Create a Pandas Excel writer using Openpyxl as the engine.
writer = pd.ExcelWriter(output_file, engine='openpyxl')

# Convert the dataframe to an Excel object.
df_grouped_sales.to_excel(writer, sheet_name='Sheet1', index=False)

# Close the Pandas Excel writer and output the Excel file.
writer.close()

print(f"集計結果を '{output_file}' に保存しました。")

# --- Apply header styling using openpyxl directly ---

# Load the workbook
workbook = openpyxl.load_workbook(output_file)
sheet = workbook['Sheet1']

# Define the light gray fill color
light_gray_fill = PatternFill(start_color='F2F2F2', end_color='F2F2F2', fill_type='solid')

# Apply the fill to the header row (row 1)
for cell in sheet[1]:
    cell.fill = light_gray_fill

# Save the modified workbook
workbook.save(output_file)

print(f"ヘッダーの背景色を適用し、'{output_file}' に保存しました。")
```

```python
import pandas as pd

# Read the Excel file into a DataFrame
df_2022 = pd.read_excel('2022_年間売上表.xlsx', sheet_name='Sheet1')

# Add a new column '売上年' with the value '2022'
df_2022['売上年'] = '2022'

# Display the first 5 rows of the DataFrame
print("First 5 rows of df_2022 with '売上年' column:")
df_2022.head()
```

```python
import pandas as pd

# Read the Excel file into a DataFrame
df_2023 = pd.read_excel('2023_年間売上表.xlsx', sheet_name='Sheet1')

# Add a new column '売上年' with the value '2023'
df_2023['売上年'] = '2023'

# Display the first 5 rows of the DataFrame
print("First 5 rows of df_2023 with '売上年' column:")
df_2023.head()
```

```python
df_combined = pd.concat([df_2022, df_2023], ignore_index=True)

print("First 5 rows of df_combined:")
df_combined.head()
```

```python
df_grouped_sales = df_combined.groupby(['商品', '売上年'])['金額（千円）'].sum().reset_index()

print("Grouped sales data (first 5 rows):")
df_grouped_sales.head()
```
